# Matrix addition

```python
from PIL import Image
import numpy as np
from ipywidgets import interact_manual
%matplotlib inline
```

## 1.1Load iamge

```python
# put your own iamges in here
A_path = './house.jpg'
B_path = './yellow.jpg'

A_image = Image.open(A_path)
B_image = Image.open(B_path)
A_image = A_image.resize((500,300))
B_image = B_image.resize((500,300))
print(A_image.size)
print(B_image.size)

display(A_image)
display(B_image)
```

## 1.2 Iamge addition

```python
A_matrix = np.array(A_image)
B_matrix = np.array(B_image)

AB = A_matrix*0.6+B_matrix*0.4
AB = Image.fromarray(AB.astype('uint8')).convert('RGB')

display(AB)
```

```python
img = Image.blend(A_image, B_image, 0.1)
#blended_img = imgA * (1 – alpha) + imgB* alpha
display(img)
```

## 1.3Drag bar to change the weight of two image


```python
def image_add(alpha):
    
    A_matrix = np.array(A_image)
    B_matrix = np.array(B_image)

    AB = A_matrix * alpha + B_matrix * (1 - alpha)
    AB = Image.fromarray(AB.astype('uint8')).convert('RGB')

    display(AB)

interact_manual(image_add, alpha=(0.0,1.0,0.01))
```

```python
A_image = Image.open('gradient.png')
B_image = Image.open('gradient2.png')
display(A_image)
display(B_image)
```

```python
A_matrix = np.array(A_image)
B_matrix = np.array(B_image)
alpha = 0.5
rotation = A_matrix * alpha + B_matrix*(1-alpha)
rotation = Image.fromarray(rotation.astype('uint8')).convert('RGB')
display(rotation)
```
