# Similartity example

```python
from ipywidgets import interact_manual
from PIL import Image
import numpy as np
```

## 1.1 load image

```python
A_image = Image.open('gradient.png')
B_image = Image.open('gradient2.png')
display(A_image)
display(B_image)
```

## 1.2 drag bar to change the weight of two image

```python
def image_add(alpha):
    
    A_matrix = np.array(A_image)
    B_matrix = np.array(B_image)
    
    rotation = A_matrix * (1-alpha) + B_matrix * alpha

    A_matrix_norm = A_matrix.flatten()/np.linalg.norm(A_matrix.flatten()) 
    rotation_norm = rotation.flatten()/np.linalg.norm(rotation.flatten()) 
    print("similarity")
    print(A_matrix_norm @ rotation_norm)
    rotation_img = Image.fromarray(rotation.astype('uint8')).convert('RGB')
    display(rotation_img)

interact_manual(image_add, alpha=(0.0,1.0,0.01))
```

# 2. similarity in image

```python
A_name = './celebrity/junkai.jpg'
B_name = './celebrity/yuan.jpg'
C_name = './celebrity/bingbing.jpg'
A_image = Image.open(A_name).resize((100,100))
B_image = Image.open(B_name).resize((100,100))
C_image = Image.open(C_name).resize((100,100))
display(A_image)
display(B_image)
display(C_image)
```

```python
A_matrix = np.array(A_image)
B_matrix = np.array(B_image)
A_matrix_norm = A_matrix.flatten()/np.linalg.norm(A_matrix.flatten()) 
B_matrix_norm = B_matrix.flatten()/np.linalg.norm(B_matrix.flatten()) 
print("%s and  %s" %(A_name, B_name))
print("Similarity:", A_matrix_norm @ B_matrix_norm)
```

```python
C_matrix = np.array(C_image)
C_matrix_norm = C_matrix.flatten()/np.linalg.norm(C_matrix.flatten()) 
print("%s and  %s" %(A_name, C_name))
print("Similarity:", A_matrix_norm @ C_matrix_norm)
```
